```python
import cv2 as cv
import numpy as np
from keras.models import model_from_json, load_model
from keras.preprocessing.image import img_to_array
from keras.preprocessing import image
import keras
from matplotlib import pyplot as plt
```

```python
detect = {0: 'Angry', 1:'Disgusted', 2:'Fearful', 3:'Happy', 4:'Neutral', 5:'sad', 6:'Suprised'}
```

```python
json_file = open('./gray_model_jsonfile.json', 'r')
load_model = json_file.read()
json_file.close()
model = model_from_json(load_model)
model.load_weights('./gray_model.weights.h5')
```

```python
cap = cv.VideoCapture(0)
```

```python
while True:
    ret, frame = cap.read()
    frame = cv.resize(frame, (1920, 1080))
    if not ret:
        break

    face_detector = cv.CascadeClassifier('./cascade_classifier/haarcascade_frontalface_default.xml')
    rgb_frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

    no_faces = face_detector.detectMultiScale(rgb_frame, scaleFactor=1.3, minNeighbors=4)

    for x, y, w, h in no_faces:
        cv.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 8)
        roi_frame = rgb_frame[y:y +h, x:x + w]
        crop_img = np.expand_dims(np.expand_dims(cv.resize(roi_frame, (48, 48)), -1), 0)

        prediction = model.predict(crop_img)
        maxidx = int(np.argmax(prediction))
        cv.putText(frame, detect[maxidx], (x, y-10), cv.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
        
    cv.imshow('detection', frame)

    if cv.waitKey(1) & 0xFF == ord('q'):
        break  # Exit the loop if 'q' key is pressed
   
cap.release()
cv.destroyAllWindows()
```
